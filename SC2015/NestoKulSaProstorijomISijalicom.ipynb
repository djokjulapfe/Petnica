```python
#Definicija funkcije
def f(x, y):
    return 20 + (x**2 - 10*np.cos(np.pi*2*x)) + (y**2 - 10*np.cos(2*np.pi*y))

#Globalne konstante
granice = np.array([[0, 5], [0, 10]]) #Granice u kojima  se trazi minimum
sigma = 0.1 #Standardna devijacija za mutiranje
population = 10 #Ukupna populacija
cutoff = 3 #Broj prezivelih u selekciji

#Racunanje fitness-a
def calcFitness(dna):
    for i in range(population):
        dna[i][2] = skf([dna[i][0], dna[i][1]])
    return dna

#Selekcija najboljih
def naturalSelection(dna):
    for i in range(cutoff, population):
        nejbr = rnd.randint(0, cutoff-1)
        dna[i][0] = dna[nejbr][0] + rnd.normal(0, sigma)
        if dna[i][0] < 0:
            dna[i][0] = 0
        if dna[i][0] > 5:
            dna[i][0] = 5
        dna[i][1] = dna[nejbr][1] + rnd.normal(0, sigma)
        if dna[i][1] < 0:
            dna[i][1] = 0
        if dna[i][1] > 10:
            dna[i][1] = 10
    return dna

#Generisanje pocetnih vrednosti
dna = rnd.rand(population, 3)
dna[:,0] = dna[:,0]*5
dna[:,1] = dna[:,1]*10

#Glavni loop
for i in range(10):
    print i
    dna = calcFitness(dna)
    dna = dna[dna[:, 2].argsort()]
    dna = naturalSelection(dna)

#Ispis
print f(dna[0][0], dna[0][1])
print dna[0][0]
print dna[0][1]
```
