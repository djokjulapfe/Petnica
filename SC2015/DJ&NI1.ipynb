```python
import matplotlib.pyplot as mp
import numpy as np
import math

# Numericki izvod funkcije f u koordinati x sa korakom h
def fp (f, x, h ):
    return (f(x+h)-f(x))/h

# Funkcija za aproksimaciju
def fun ( x ):
    #return np.exp(-2*x)
    return np.sin(x)

h = 0.1 #Korak ojlera
t = 30 #Vremenski interval
x = np.zeros(t/h) #x koordinate (0.0, 0.1, 0.2, ...)
y = np.zeros(t/h) #Ojlerova aproksimacija
y1 = np.zeros(t/h) #Analiticko resenje
y[0] = 0 #Pocetni uslov
y1[0] = fun(0)

for i in range (1, int(t/h)): #svaki korak je pretstavljen kao h*i
    x[i] = i*h;
    y[i] = y[i-1] + h * fp(fun, h*i, h)
    y1[i] = fun(i*h)
    
mp.plot(x,y,'-r',x,y1,'-b') # Plotovanje numericke (crveno) i analiticke (plavo)
mp.show()
```

```python
import matplotlib.pyplot as mp
import numpy as np
import math

G = 6.67e-11
m1 = 2e30
m2 = 2e24

def F( r1, r2 ):
    fx = -G*m1*m2/(r1[0]-r2[0])**3*np.abs(r1[0]-r2[0])
    fy = -G*m1*m2/(r1[1]-r2[1])**3*np.abs(r1[1]-r2[1])
    return np.array([fx, fy])

h = 3600
t = 3600*365

r1 = np.zeros((t/h,2))
r1[0] = np.array([0, 0])
r2 = np.zeros((t/h,2))
r2[0] = np.array([1.5e11, 0])
v1 = np.zeros((t/h,2))
v1[0] = np.array([0, 0])
v2 = np.zeros((t/h,2))
v2[0] = np.array([0, 3e4])

for i in range(1, int(t/h)):
    
    k1v1 = F(r1[i], r2[i])/m1
    k1v2 = F(r1[i], r2[i])/m2
    k1r1 = v1[i]
    k1r2 = v2[i]
    
    k2v1 = F(r1[i]+k1r1*h/2, r2[i]+k1r2*h/2)/m1
    k2v2 = F(r1[i]+k1r1*h/2, r2[i]+k1r2*h/2)/m2
    k2r1 = v1[i] + k1r1*h/2
    k2r2 = v2[i] + k1r2*h/2
    
    k3v1 = F(r1[i]+k1r1*h/2, r2[i]+k1r2*h/2)/m1
    k3v2 = F(r1[i]+k1r1*h/2, r2[i]+k1r2*h/2)/m2
    k3r1 = v1[i] + k2r1*h/2
    k3r2 = v2[i] + k2r2*h/2
    
    k4v1 = F(r1[i]+k1r1*h, r2[i]+k1r2*h)/m1
    k4v2 = F(r1[i]+k1r1*h, r2[i]+k1r2*h)/m2
    k4r1 = v1[i] + k1r1*h
    k4r2 = v2[i] + k1r2*h
    
    kv1 = (k1v1 + 2*k2v1 + 2*k3v1 + k4v1)/6
    kv2 = (k1v2 + 2*k2v2 + 2*k3v2 + k4v2)/6
    kr1 = (k1r1 + 2*k2r1 + 2*k3r1 + k4r1)/6
    kr2 = (k1r2 + 2*k2r2 + 2*k3r2 + k4r2)/6
    
    r1[i] = r1[i-1] + h*kr1
    r2[i] = r2[i-1] + h*kr2
    v1[i] = v1[i-1] + h*kv1
    v2[i] = v2[i-1] + h*kv2
    
mp.plot(r1[0],r1[1],'-r',r2[0],r2[1],'-b') # Plotovanje numericke (crveno) i analiticke (plavo)
mp.show()
```
